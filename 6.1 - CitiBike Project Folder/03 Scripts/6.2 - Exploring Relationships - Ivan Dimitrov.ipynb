## Exploring Correlations

```python
# Checking Data types
df_bikes.dtypes
```

```python
# Creating a subset
df_sub = df_bikes[['day_of_week','trip_duration_in_minutes','subscriber','year_of_birth','gender']]
```

```python
# Checking the subset
df_sub
```

```python
# Encoding the subset columns from objects to integers
day_mapping = {'Mon': 0, 'Tue': 1, 'Wed': 2, 'Thu': 3, 'Fri': 4, 'Sat': 5, 'Sun': 6}
subscriber_mapping = {'Subscriber': 1, 'Non-Subscriber': 0}
gender_mapping = {'Male': 1, 'Female': 2, 'Unknown': 0}
```

```python
df_sub['day_of_week_encoded'] = df_sub['day_of_week'].map(day_mapping)
df_sub['subscriber_encoded'] = df_sub['subscriber'].map(subscriber_mapping)
df_sub['gender_encoded'] = df_sub['gender'].map(gender_mapping)
```

```python
df_sub
```

```python
# Dropping the unnecessary columns
df_sub = df_sub.drop(columns = ['day_of_week','subscriber','gender'])
```

```python
# Checking the results
df_sub
```

```python
df_sub.corr()
```

## Correlation Matrix heatmap with matplotlib

```python
# Creating a correlation heatmap using matplotlib

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

There seems to be a medium to strong correlation between "gender - subscriber". The very negative relationship between "subscriber - trip duration" is also very interesting.

## Creating Correlation heatmap using seaborn

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

This heatmap using seaborn is more explanatory. It shows visually the % of correlation. It solidifies the relationships between the variables by showing the percentage on top of color. We can still see the two groups that are positively and negatively relatable (gender - subscriber) and (subscriber - trip duration).

## Categorical Plots

```python
df_sub.describe()
```

```python
# Creating a categorical plot
sns.histplot(df_sub['trip_duration_in_minutes'], bins = 20, kde = True)
```

From our categorical plot here we can see that the most of the users are using Citibikes for the duration between 300 minutes and 600 minutes (5 - 10 hours). Which indicates that the users are probably using it for going to work or during weekends.

```python
# Creating categories for the trip duration
df_sub.loc[df_sub['trip_duration_in_minutes'] < 300, 'trip duration category'] = 'Short trip'
```

```python
df_sub.loc[(df_sub['trip_duration_in_minutes'] >= 300) & (df_sub['trip_duration_in_minutes'] < 600), 'trip duration category'] = 'Middle trip'
```

```python
df_sub.loc[df_sub['trip_duration_in_minutes'] >= 600, 'trip duration category'] = 'Long trip'
```

```python
df_sub['trip duration category'].value_counts(dropna = False)
```

```python
# Creating a categorical plot in seaborn using the trip duration categories created above

sns.set(style="ticks")
g = sns.catplot(x="day_of_week_encoded", y="trip_duration_in_minutes", hue="trip duration category", data=df_sub)
```

- We can see that most of the trips are "Long trips" and more than 600 minutes is more than 10 hours, which leads me to the conclusion that people are using the bikes on a weekly bases. They rent the bike for the entire week. Which in itself leads me to the conclusion that most of the users are people who use the NYCitiBikes for commute reasons.
