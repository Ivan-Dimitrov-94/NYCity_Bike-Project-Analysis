### Wrangling

```python
# Checking the columns and basic info
df_bikes.info()
```

```python
# Dropping columns I don't need for the potential questions and analysis
df_bikes_clean = df_bikes.drop(columns = ['trip_id','bike_id','start_station_id','end_station_id','start_hour'])
```

```python
# Checking the cleaned data
df_bikes_clean
```

```python
# Checking data info after dropped columns
df_bikes_clean.info()
```

```python
# Changing the type of data to lower the memory usage of the file
df_bikes_clean['trip_duration']=df_bikes_clean['trip_duration'].astype('int16')
```

```python
# Changing gender to a string value
df_bikes_clean['gender']=df_bikes_clean['gender'].astype('str')
```

```python
# Checking data info after type change
df_bikes_clean.info()
```

```python
# Changing the "gender" column to "int8" so we can perform the transformation of (0,1,2) to (unknown,male,female)
df_bikes_clean['gender']=df_bikes_clean['gender'].astype('int8')
```

```python
# Changing the gender variables to actual genders instead of numbers
df_bikes_clean['gender']=df_bikes_clean['gender'].replace([0,1,2],['Unknown','Male','Female'])
```

```python
# Reverting "gender" column back to a string
df_bikes_clean['gender']=df_bikes_clean['gender'].astype('str')
```

```python
# Checking the result
df_bikes_clean.head()
```

```python
# Showing the descriptive stats of the data set
df_bikes_clean.describe()
```

```python
# Replacing the NaN values inside column "birth_year" with the Mode of the data
df_bikes_new = df_bikes_clean.fillna("1978.000000")
```

```python
# Checking the data after the changes
df_bikes_new.info()
```

```python
# Changing "birth_year" column to a float so we can input the Mode value in the NaN place
df_bikes_new['birth_year']=df_bikes_new['birth_year'].astype('float64')
```

```python
# Checking the data
df_bikes_new.info()
```

```python
# Visualizing the change
df_bikes_new.describe()
```

```python
# Reverting back the "birth_year" column to a string in order to see the years as whole numbers instead of continuous
df_bikes_new['birth_year'] = df_bikes_new['birth_year'].astype('int16')
```

```python
# Checking the data after all the changes
df_bikes_new
```

```python
# Renaming some columns for ease of use
df_bikes_new.rename(columns = {'weekday':'day_of_week'}, inplace = True)
df_bikes_new.rename(columns = {'start_time':'start_of_ride'}, inplace = True)
df_bikes_new.rename(columns = {'end_time':'end_of_ride'}, inplace = True)
df_bikes_new.rename(columns = {'trip_duration':'trip_duration_in_minutes'}, inplace = True)
df_bikes_new.rename(columns = {'birth_year':'year_of_birth'}, inplace = True)
```

```python
# Checking the new dataset
df_bikes_new
```

```python
# Removing data for people born before 1933, because I do not consider bike riding data for people aged above 80 years old as relevant
df_bikes_latest = df_bikes_new[df_bikes_new['year_of_birth'] >= 1938]
```

```python
# Checking the filtered data
df_bikes_latest.info()
```
